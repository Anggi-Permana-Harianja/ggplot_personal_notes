```python
from plotnine.data import economics, mpg
from plotnine import ggplot, aes, geom_line, geom_point, geom_smooth
```

```python
(
    ggplot(economics)
    + aes(x='date', y='pop')
    + geom_line()
)
```

```python
(
    ggplot(data=mpg,
           mapping=aes(x='displ', y='hwy'))
    + geom_point()
    + geom_smooth(method='loess')
)
```
